```python
import matplotlib.pyplot as plt
import time

# Read the recorded states from the text file
with open("bitonic_states_v1.txt", "r") as f:
    lines = f.readlines()

states = []
labels = []

# Parse the lines
for line in lines:
    label, values_str = line.split(":")
    values = list(map(int, values_str.strip().split()))
    states.append(values)
    labels.append(label)

# Show each state as an individual chart
for i, values in enumerate(states):
    plt.figure(figsize=(10, 5))
    plt.bar(range(len(values)), values, color='skyblue')
    plt.title(f"Step {i + 1}: {labels[i]}")
    plt.xlabel("Index")
    plt.ylabel("Value")
    plt.tight_layout()
    plt.savefig(f"v1_pictures/step_{i + 1}.png")
    plt.show()
    
    # Optional: wait before next chart appears (comment out if using in Jupyter)
    time.sleep(0.01)


# Read the recorded states from the text file
with open("bitonic_states_v2.txt", "r") as f:
    lines = f.readlines()

states = []
labels = []

# Parse the lines
for line in lines:
    label, values_str = line.split(":")
    values = list(map(int, values_str.strip().split()))
    states.append(values)
    labels.append(label)

# Show each state as an individual chart
for i, values in enumerate(states):
    plt.figure(figsize=(10, 5))
    plt.bar(range(len(values)), values, color='skyblue')
    plt.title(f"Step {i + 1}: {labels[i]}")
    plt.xlabel("Index")
    plt.ylabel("Value")
    plt.tight_layout()
    plt.savefig(f"v2_pictures/step_{i + 1}.png")
    plt.show()
    
    # Optional: wait before next chart appears (comment out if using in Jupyter)
    time.sleep(0.01)

# Read the recorded states from the text file
with open("bitonic_states_v3.txt", "r") as f:
    lines = f.readlines()

states = []
labels = []

# Parse the lines
for line in lines:
    label, values_str = line.split(":")
    values = list(map(int, values_str.strip().split()))
    states.append(values)
    labels.append(label)

# Show each state as an individual chart
for i, values in enumerate(states):
    plt.figure(figsize=(10, 5))
    plt.bar(range(len(values)), values, color='skyblue')
    plt.title(f"Step {i + 1}: {labels[i]}")
    plt.xlabel("Index")
    plt.ylabel("Value")
    plt.tight_layout()
    plt.savefig(f"v3_pictures/step_{i + 1}.png")
    plt.show()
    
    # Optional: wait before next chart appears (comment out if using in Jupyter)
    time.sleep(0.01)
```
